```python
import os
import geopandas as gpd 
import pandas as pd
from shapely import box
```

```python
data_dir = "../data/"
```

```python
cglrsed_attrs = pd.read_parquet(os.path.join(data_dir, "deafrica_waterbodies_cglrsed_update.parquet"), columns=["uid", "GLRSED_Reservoir", "GLRSED_Name"])
hydroatlas_lakeatlas_attrs = pd.read_parquet(os.path.join(data_dir, "deafrica_waterbodies_lakeatlas_update.parquet"), columns=["uid", "LakeATLAS_Reservoir", "LakeATLAS_Name"])
realsat_attrs = pd.read_parquet(os.path.join(data_dir, "deafrica_waterbodies_realsat_update.parquet"), columns=["uid", "ReaLSAT_Reservoir"])
```

```python
# Load the DE Africa Waterbodies Historical Extent Product.
deafrica_waterbodies_fp =  os.path.join(data_dir, "waterbodies.parquet")
deafrica_waterbodies = gpd.read_parquet(deafrica_waterbodies_fp)
print(f"Loaded {len(deafrica_waterbodies)} DE Africa waterbodies")
```

```python
# Merge the attribute tables
attrs = cglrsed_attrs.merge(hydroatlas_lakeatlas_attrs, on="uid").merge(realsat_attrs, on="uid")
attrs
```

```python
deafrica_waterbodies = deafrica_waterbodies.merge(attrs, on="uid")
deafrica_waterbodies
```

```python
deafrica_waterbodies.to_parquet(os.path.join(data_dir, "deafrica_waterbodies_combined_attrs.parquet"))
```
